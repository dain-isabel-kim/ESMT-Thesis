```python
import pandas as pd
import numpy as np
import networkx as nx
from matplotlib.pyplot import figure
attributesInput = pd.read_csv('../inputdata/AttributesF.csv')
emailDataInput = pd.read_csv('../inputdata/FinalEdata.csv')
conflictsInput = pd.read_csv('../inputdata/TC2.csv')
```

```python
attributes = attributesInput
print(attributes)
```

```python

conflicts = conflictsInput.set_index(['Source'])
conflicts = conflicts.stack().to_frame()
```

```python
conflicts = conflicts.rename_axis(['src', 'dest']).reset_index()
```

```python
conflicts = conflicts.rename(columns={0: 'type'}) 
```

```python
conflicts = conflicts.astype({'type': 'int16'})
conflicts = conflicts.astype({'dest': 'int32'})

print(conflicts)
```

```python
conflicts = conflicts.merge(attributesInput, how='left', left_on='src', right_on='ID')
```

```python
conflicts = conflicts.drop(columns=['ID', 'Gender', 'Age', 'Tenure_Full', 'Supervisor_T1', 'Hierarchical Level'])
conflicts = conflicts.rename(columns={'Department': 'dep_src'})
```

```python
conflicts = conflicts.merge(attributesInput, how='left', left_on='dest', right_on='ID')
conflicts = conflicts.drop(columns=['ID', 'Gender', 'Age', 'Tenure_Full', 'Supervisor_T1', 'Hierarchical Level'])
conflicts = conflicts.rename(columns={'Department': 'dep_dest'})
conflicts = conflicts.astype({'dep_src': 'int16', 'dep_dest': 'int16'})
conflicts = conflicts.loc[conflicts['type'] == 1]
print(conflicts)
```

```python
target_dept = 2
dept = conflicts.loc[ ((conflicts['dep_src'] == target_dept) | (conflicts['dep_dest'] == target_dept)) ]
```

```python
G = nx.from_pandas_edgelist(dept, 'src', 'dest', create_using=nx.MultiDiGraph)
colors = []
for node in G:
    if attributes.loc[node-1, 'Department'] == target_dept:
        colors.append('red')
    else:
        colors.append('lightblue')
```

```python
figure(figsize=(40, 40))
nx.draw_spring(G, with_labels=True, node_color=colors, node_size=1500, font_size=35)
```
